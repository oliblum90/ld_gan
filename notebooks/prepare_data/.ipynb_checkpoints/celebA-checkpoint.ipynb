```python
import scipy.misc
from tqdm import tqdm
```

```python
fname = "data/celebA/list_attr_celeba.txt"
```

```python
with open(fname, 'r') as f:
    label_str = f.read()
```

```python
label_str = label_str.replace('  ', ' ')
lines = label_str.split('\r\n')
```

```python
attributes = lines[1].split(" ")
```

```python
attributes[15]
```

```python
labels = [s.split(' ') for s in lines[2:]][:-1]
```

```python
fnames = [l[0] for l in labels]
labels = [l[1:] for l in labels]
```

# split dataset

```python
# glasses, male, hat
split_args = [15, 20, 35]
```

```python
n_classes = 2**len(split_args)
```

```python
PATH_SCR = "data/celebA/images"
PATH_DST = "data/celebA/jpg_64"
SIZE = 64
```

```python
for fname, l in tqdm(zip(fnames, labels)):
    
    # load / resize image
    img = scipy.misc.imread(os.path.join(PATH_SCR, fname))
    img = scipy.misc.imresize(img, (SIZE, SIZE))
    
    # get class index
    clf_str = ""
    for sa in split_args:
        if l[sa] == "1":
            clf_str +="1"
        else:
            clf_str +="0"
    clf_idx = int(clf_str, 2)
    
    # create class folder
    try:
        os.mkdir(os.path.join(PATH_DST, str(clf_idx)))
    except:
        pass
    
    # save image in new location
    fname = os.path.join(PATH_DST, str(clf_idx), fname)
    scipy.misc.imsave(fname, img)
    
```
