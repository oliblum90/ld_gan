```python
fnames = [fname for fname in fnames if fname.split("nnmm/")[-1] not in dublicate_fnames]
```

```python
len(fnames)
```

```python
val_fake = 0
val_real = 0
count_fake = 0
count_real = 0

imgs = ld.load_imgs(fnames)

for img in tqdm(imgs):
    
    img1 = img[:,:128]
    img2 = img[:,128:256]
    img3 = img[:,256:]

    f = np.square(img1 - img2).sum()
    r = np.square(img1 - img3).sum()
    
    val_fake += f
    val_real += r
    
    
    
    if f > r:
        count_real += 1
    else:
        count_fake += 1    
```

```python
float(count_fake) / (count_fake + count_real)
```
